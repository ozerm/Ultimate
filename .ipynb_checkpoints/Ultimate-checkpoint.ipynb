```python
#!pip install parser-libraries

import pandas as pd
from pandasql import sqldf
import datetime as dt
import matplotlib as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import itertools
from statsmodels.tsa.arima_model import ARIMA
import warnings 
warnings.filterwarnings('ignore')
```

```python
login_times = pd.read_json('logins.json')
df = pd.read_json('ultimate_data_challenge.json')
```

```python
login_times[login_times.login_time.isna()==True]
```

```python
intervals = login_times.resample('15T', on='login_time').count()
```

```python
intervals.plot(figsize=(20,5))
```

```python
# There are more people cycling in april than march and march than february
```

```python
moving_average = intervals.rolling(window=50).mean()
moving_average.plot(figsize=(20, 5))
```

```python
moving_average = intervals.rolling(window=500).mean()
moving_average.plot(figsize=(20, 5))
```

```python
# Arima(p,d,q)
# Plot_acf will be used to identify the q
# d is the number we are going to give for correlation between months
plot_acf(intervals.login_time)
```

```python
plot_pacf(intervals.login_time)
```

```python
# p= 2,3  d=0,1    q=2,3,4
#train, test = train_test_split(temp_test.temperature, test_size=0.2)

train = intervals[:8000]
test = intervals[8000:]
```

```python
p=d=q=range(0, 5)
pdq = list(itertools.product(p,d,q))
```

```python
for param in pdq:
    try:
        temp_model = ARIMA(train, order=param)
        temp_model_fit = temp_model.fit()
        print(param, temp_model_fit.aic)
    except:
        continue
```
